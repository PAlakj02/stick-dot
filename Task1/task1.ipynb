Text to 3D Model (SketchUp)


```python
# Install Trimesh (lightweight 3D library)
!pip install trimesh

import trimesh
import numpy as np

# Create a simple house (cube + pyramid roof)
def create_house():
    # Base (cube)
    base = trimesh.creation.box(extents=[10, 10, 5])
    # Roof (pyramid)
    vertices = np.array([[0, 0, 5], [10, 0, 5], [10, 10, 5], [0, 10, 5], [5, 5, 10]])
    faces = np.array([[0, 1, 4], [1, 2, 4], [2, 3, 4], [3, 0, 4]])
    roof = trimesh.Trimesh(vertices=vertices, faces=faces)
    # Combine
    house = base + roof
    return house

# Generate and export
house = create_house()
house.export('house.obj')
print("3D house model saved as 'house.obj'")
```

Text to Animation

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Set up the figure
fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 100)
line, = ax.plot(x, np.sin(x))

# Animation function
def update(frame):
    line.set_ydata(np.sin(x + frame/10))
    return line,

# Animate
ani = FuncAnimation(fig, update, frames=100, interval=50)
plt.title("Sine Wave Animation")
plt.show()

# Save as GIF (optional)
ani.save('sine_wave.gif', writer='pillow', fps=20)
```

Text to editable infographs


```python
!pip install plotly

import plotly.express as px

# Data from text (e.g., "Show sales for A=100, B=200, C=150")
data = {"Category": ["A", "B", "C"], "Sales": [100, 200, 150]}

# Create interactive bar chart
fig = px.bar(data, x='Category', y='Sales', title="Sales Infographic")
fig.show()

# Save as HTML (editable)
fig.write_html("infographic.html")
```
